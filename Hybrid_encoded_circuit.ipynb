```python
from qiskit import *
%matplotlib inline
from math import pi,sqrt
```

```python
#Encoded length = 15
circ = QuantumCircuit(15)
all_qubits = [i for i in range(15)]
theta = qiskit.circuit.Parameter('theta')
        
circ.h(all_qubits)
circ.barrier()
circ.ry(theta, all_qubits)        
circ.measure_all()

circ.draw('mpl')
```
